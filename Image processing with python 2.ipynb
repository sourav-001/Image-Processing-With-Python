### if i have a whole bunch of images e.g "Images" folder how can i read all images, or say how can i read subfolders?  for this i will use glob 
### glob finds all path names


```python
import cv2
import glob

# path name
path = "Images/*"

for file in glob.glob(path):
    # print file names one by one that are in the path folders 
    print(file)
    
    # reading file in cv2
    a = cv2.imread(file)
    # print
    print(a)
    
    # printing original iamge
    cv2.imshow("Original image", a)
    cv2.waitKey(2000)
    cv2.destroyAllWindows()
    
    
    # convert color of images BGR to RGB
    c = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    
    # show image
    cv2.imshow("Converted image", c)
    
    # i put the value 2000 so that i dont have to close the window manually 
    cv2.waitKey(2000)
    cv2.destroyAllWindows()
```
