An image is nothing but an array of numbers or pixels and each pixel is nothing but a number.
An image is a numpy array. And for color images at each pixel the values are red, green, blue so this is how the pixel values
are defined and in , and with python you can think these as four values red, green, blue and alpha.
Alpha defines the transparency of your image. 

```python
# importing nenecessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing skimage library
import skimage.io as io
# for converting in float
from skimage import img_as_float

# reading image
my_image = io.imread("Images/1_jpeg.jpg")

# printing the image
# here i am going to print the image insted of showing
# here i can see that an image is nothing but a numbers array (here is numpy array)

print(my_image)

# now i am printing the maximum and the minimum number in the nu,py array

print(my_image.min(), my_image.max())
```

so i see that the array is started from 0 to 255


```python
# show my image
plt.imshow(my_image)
```

```python
# creating a random image with help of numpy

random_image = np.random.random([500, 500])
# showing the image basically ploting the image

plt.imshow(random_image)

# here again i am going to print the image

print(random_image)

# min and max

print(random_image.min(), random_image.max())
```

here( here means in the previous block) i get an array but not an integer array, insted of that i get an array with floating number

```python
# converting my_image to float
# previously i see that the array i get it was an integre array

my_image_in_float = img_as_float(my_image)
print(my_image_in_float)
# min and max
print(my_image_in_float.min(), my_image_in_float.max())
```

again another thing i see that if i use "img_as_float" i get 0.0 to 1.0
but if i use numpy i get float number but range is not between 0.0 to 1.0

```python
# here i am reducing all the values with the image by half

reduce_my_image = my_image * 0.5
print(reduce_my_image)
plt.imshow(reduce_my_image)
```

```python
# creating a new numpy array that has all red overlap in my image

my_new_image = my_image
my_new_image[10:200, 10:200, :] = [255,0,0]
plt.imshow(my_new_image)
```

```python
my_new_image = my_image
my_new_image[200:400, 200:400, :] = [0,255,0]
plt.imshow(my_new_image)
```

```python
my_new_image = my_image
my_new_image[400:600, 400:600, :] = [0,0,255]
plt.imshow(my_new_image)
```
