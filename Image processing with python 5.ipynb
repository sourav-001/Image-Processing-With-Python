# image processing using scikit image



```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing skimage
# io is for read and write images in various formats
import skimage.io as io

# for rescale and resize
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
# read image
img = io.imread("Images/AVengers/spiderman_1.jpg")

# read image as grey
img_grey = io.imread("Images/AVengers/spiderman_1.jpg", as_gray=True)
```

```python
# show original image
plt.imshow(img)
```

```python
# show grey image of the original image
#plt.imshow(img_grey)

io.imshow(img_grey)
```

```python
# rescale image
rescale_grey_image = rescale(img_grey, 1.0/4.0, anti_aliasing=True)

# resize image
# resize the image to match a certain size
# performs interpolation to up-size or down-size N-dimensional images
resize_grey_image = resize(img_grey, (200, 200), anti_aliasing=True)

# downscale using local mean
# down sample N-dimensional image by local mean
downscale_grey_image = downscale_local_mean(img_grey, (4,2))
```

```python
img_grey.shape
```

```python
# show rescale of the grey image
#plt.imshow(rescale_grey_image)
io.imshow(rescale_grey_image)
```

```python
# show resized grey scale image
io.imshow(resize_grey_image)
```

```python
# show downscale image 
io.imshow(downscale_grey_image)
```

```python
# few ways of detection for edges
from skimage.filters import roberts, sobel, scharr, prewitt
```

```python
# cannay
from skimage.feature import canny
# canny is not only judt do edge detection
# canny can do many more things (many more operations)
# it does noise reduction, gradient calculation, edge tracking
```

```python
edge_roberts = roberts(img_grey, mask=None)
#io.imshow(edge_roberts)
plt.imshow(edge_roberts, cmap="gray")
```

```python
io.imshow(edge_roberts)
```

```python
# nuron image load
nuron_img = io.imread("Images/Cell_nuron/nuron-1.png", as_gray=True)
```

```python
edge_roberts_nuron_img = roberts(nuron_img, mask=None)
edge_sobel_nuron_img = sobel(nuron_img)
edge_sacharr_nuron_img = scharr(nuron_img)
edge_prewitt_nuron_img = prewitt(nuron_img)


# plotting

fig, axes = plt.subplots(nrows=3, ncols=2, sharex=True, sharey=True, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(nuron_img, cmap= plt.cm.gray)
ax[0].set_title("Original Image")

ax[1].imshow(edge_roberts_nuron_img, cmap= plt.cm.gray)
ax[1].set_title("Roberts Edge Detection")

ax[2].imshow(edge_sobel_nuron_img, cmap= plt.cm.gray)
ax[2].set_title("Sobel Edge Detection")

ax[3].imshow(edge_sacharr_nuron_img, cmap= plt.cm.gray)
ax[3].set_title("Sacharr Edge Detection")

ax[4].imshow(edge_prewitt_nuron_img, cmap= plt.cm.gray)
ax[4].set_title("Prewitt Edge Detection")

for a in ax:
    a.axis("off")

plt.tight_layout()
plt.show()    
```

```python
# amount of edges in sigma
edge_canny_nuron_img = canny(nuron_img, sigma=7)
plt.imshow(edge_canny_nuron_img)
```
