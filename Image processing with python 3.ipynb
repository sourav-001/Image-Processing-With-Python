# image processiong using pillow in python

```python
# importing some necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pillow library is o library for basic imgae processing like image croping,
# image rotation, changing the color spaces etc

# now i import pillow library
# i have imported full pillow library, but i do not need full library here 
# so i have to import Image for pillow
# PIL or image enables us to load our images
import PIL
from PIL import Image
```

```python

img = Image.open("Images/1_jpeg.jpg")
# this image is not in numpy array, in pillow if needed we have to convert in numpy array

print("Type of image is:", type(img))

# printing image format
print(img.format)

# printing image mode
print(img.mode)

# image size width and height
print(img.size)

# showing image
img.show()

# resizing imge
small_img = img.resize((700, 700))
# for save image
#small_image.save("Images/small_image.jpg")
#small_img.show()

# thumbnails
img.thumbnail((500, 400))
img.save("Images/1_jpeg_Image_thumbnail.jpg")
print(img.size)
```

```python
# croping image

# load a new image
img_1 = Image.open("Images/shaktiman.jpg")

# ploting the image
#plt.imshow(img_1)
#img_1.show()
# here i cropped the image
# (x1, y1, x2, y2)
# (x1 and x2 for horizontal ) (y1 and y2 for vertical)

cropped_img = img_1.crop((250, 110, 650, 580))
#cropped_img.show()
#plt.imshow(cropped_img)

# now plotting both images original and cropped
# fig 1 original image
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
original_img_plot = plt.imshow(img_1)
ax.set_title('Original Image')


# fig 2 cropped image
ax = fig.add_subplot(1, 2, 2)
cropped_img_plot = plt.imshow(cropped_img)
ax.set_title('Cropped Image')

# save the croped image
cropped_img.save("Images/Croped_shaktiman_img.jpg")
```

```python
# copy images


# load two new images
image_1 = Image.open("Images/Thor_1.jpg")
image_2 = Image.open("Images/shaktiman.jpg")

# printing image sizes
print(f"Image_1 {image_1.size}, Image_2 {image_2.size}")

# now plotting both images image_1 and image_2
# fig 1 original image
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
img1_plot = plt.imshow(image_1)
ax.set_title('Image 1')


# fig 2 cropped image
ax = fig.add_subplot(1, 2, 2)
img2_plot = plt.imshow(image_2)
ax.set_title('Image 2')

# copy image_1
image_1_copy =  image_1.copy()

# paste image_2 into image_1_Copy
image_1_copy.paste(image_2, (10, 10))

image_1_copy.show()
```

```python
plt.imshow(image_1_copy)

# save the image

image_1_copy.save("Images/copy_paste_image.jpg")
```

```python
# image rotation
img = Image.open("Images/Croped_shaktiman_img.jpg")

# rotate image 90 degree

img90 = img.rotate(90)
img90.save("Images/rotate_90.jpg")
rotate_img = Image.open("Images/rotate_90.jpg")
rotate_img.show()
```

```python
# rotate 45 degree
img45 = img.rotate(45)
#img45.show()
plt.imshow(img45)
```

```python
# rotate 45 degree nut keep want to edges
img_45 = img.rotate(45, expand=True)

#img_45.show()
plt.imshow(img_45)
```

```python
img_180 = img.rotate(180, expand=True)

#img_45.show()
plt.imshow(img_180)
```

```python
# flip image 
img_thor = Image.open("Images/Thor_2.jpg")
# original image
plt.imshow(img_thor)
```

```python
# flip image left to right
img_flip_L_R = img_thor.transpose(Image.FLIP_LEFT_RIGHT)
# plot image
plt.imshow(img_flip_L_R)
```

```python
# flip image top to bottom

img_flip_R_L = img_thor.transpose(Image.FLIP_TOP_BOTTOM)

# plot image
plt.imshow(img_flip_R_L)
```

```python
# change to grey scale

grey_image_thor = img_thor.convert("L")

#plt.imshow(grey_image_thor)
grey_image_thor.show()
```

```python
# grey with some thing 

grey_1 = img_thor.convert("1")

grey_1.show()
```

```python
# glob

import glob

path = "Images/*"
for file in glob.glob(path):
    load_file = Image.open(file)
    rotate_45d = load_file.rotate(45, expand=True)
    rotate_45d.save(file+"_rotaded45.png")
```
