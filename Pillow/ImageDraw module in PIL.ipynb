#### Draws a bitmap (mask) at the given position, using the current fill color for the non-zero portions. The bitmap should be a valid transparency mask (mode “1”) or matte (mode “L” or “RGBA”). This is equivalent to doing image.paste(xy, color, bitmap). To paste pixel data into an image, use the paste() method on the image itself.

***
`ImageDraw.bitmap(xy, bitmap, fill=None)`
***

```python
# open image
# load base imnage

image21 = Image.open("Images/voldemort_0.jpg")
image22 = Image.open("Images/voldemort_0.jpg")

# create drawing context
draw_21 = ImageDraw.Draw(im=image21)
bmsz = (image21.width // 16 - 10, image21.height // 16 -10)
colors = list(ImageColor.colormap.keys())

for x in range(16):
    for y in range(16):
        bm = Image.new("L", bmsz)
        draw_inner = ImageDraw.Draw(bm)
        draw_inner.ellipse(
            [(0, 0), bm.size],
            fill= y * 16 + x
            ) # (y * 16 +x ) varies in range(0, 256)
        del draw_inner # destroy inner draw context
        pos = [
            ((bmsz[0] + 10) * x + 10,
            (bmsz[1] + 10) * y + 10)]
        # pos is pixel values of bm is used as mask to fill
        draw_21.bitmap(pos, bm, fill=colors[(y * 16 + x) % len(colors)])

# destroy drawing context
del draw_21

# save image
image21.save("Images/Image-Result/bitmap_1.png")

# ploting both images
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
ax.set_title("Original Image")
plot_image21 = plt.imshow(image22)

ax = fig.add_subplot(1, 2, 2)
ax.set_title("Image with bitmap")
plot_image21_bitmap = plt.imshow(image21)
```

#### Draws a polygon. The polygon outline consists of straight lines between the given coordinates, plus a straight line between the last and the first coordinate

***
`ImageDraw.polygon(xy, fill=None, outline=None)`
***

1. xy – Sequence of either 2-tuples like [(x, y), (x, y), ...] or numeric values like [x, y, x, y, ...].
2. outline – Color to use for the outline.
3. fill – Color to use for the fill.

```python
# for calculating bounding box need to import ImapgePath module
from PIL import ImagePath 

ptc = 7
xy = [
    ((math.cos(th) + 1) * 120,
     (math.sin(th) + 1) * 90)
    for th in [i * (2 * math.pi) / ptc for i in range(ptc)]
    ]  # not closed (i.e., end != start)

bbox = ImagePath.Path(xy).getbbox()  # calculate bounding box
size = list(map(int, map(math.ceil, bbox[2:])))

# demo 1 - polygon with outline
image44 = Image.new("RGB", size, "#f9f9f9")  # create new Image
draw_44 = ImageDraw.Draw(image44)  # create drawing context
draw_44.polygon(xy, fill="#eeeeff", outline="blue")  # draw polygon with outline
del draw_44  # destroy drawing context
image44.save("Images/Image-Result/ImageDraw_polygon_line_01.jpg")

# demo 2 - polygon and line with the same xy
image45 = Image.new("RGB", size, "#f9f9f9")  # create new Image
draw_45 = ImageDraw.Draw(image45)  # create drawing context
draw_45.polygon(xy, fill="#eeeeff")  # draw polygon without outline
draw_45.line(xy, fill="blue", width=5)  # draw line
del draw_45  # destroy drawing context
image45.save("Images/Image-Result/ImageDraw_polygon_line_02.jpg")
```

#### Draws a rounded rectangle.

***
`ImageDraw.rounded_rectangle(xy, radius=0, fill=None, outline=None, width=1)`
***

1. xy – Two points to define the bounding box. Sequence of either [(x0, y0), (x1, y1)] or [x0, y0, x1, y1]. The second point is just outside the drawn rectangle.
2. radius – Radius of the corners.
3. outline – Color to use for the outline.
4. fill – Color to use for the fill.
5. width – The line width, in pixels.

#### Draws the string at the given position.

***
`ImageDraw.text(xy, text, fill=None, font=None, anchor=None, spacing=4, align='left', direction=None, features=None, language=None, stroke_width=0, stroke_fill=None, embedded_color=False)`
***

1. xy – The anchor coordinates of the text.
2. text – String to be drawn. If it contains any newline characters, the text is passed on to multiline_text().
3. fill – Color to use for the text.
4. font – An ImageFont instance.
5. anchor – The text anchor alignment. Determines the relative location of the anchor to the text. The default alignment is top left. See Text anchors for valid values. This parameter is ignored for non-TrueType fonts.
6. spacing – If the text is passed on to multiline_text(), the number of pixels between lines.
7. align – If the text is passed on to multiline_text(), "left", "center" or "right". Determines the relative alignment of lines. Use the anchor parameter to specify the alignment to xy.
8. direction – Direction of the text. It can be "rtl" (right to left), "ltr" (left to right) or "ttb" (top to bottom). Requires libraqm.
9. features – A list of OpenType font features to be used during text layout. This is usually used to turn on optional font features that are not enabled by default, for example "dlig" or "ss01", but can be also used to turn off default font features, for example "-liga" to disable ligatures or "-kern" to disable kerning. To get all supported features, see OpenType docs. Requires libraqm.
10. language – Language of the text. Different languages may use different glyph shapes or ligatures. This parameter tells the font which language the text is in, and to apply the correct substitutions as appropriate, if available. It should be a BCP 47 language code. Requires libraqm.
11. stroke_width – The width of the text stroke.
12. stroke_fill – Color to use for the text stroke. If not given, will default to the fill parameter.
13. embedded_color – Whether to use font embedded color glyphs (COLR, CBDT, SBIX).

```python
# centered text

def center_text(output_path):
    width, height = (400, 400)
    image101 = Image.new(mode="RGB", size=(width, height), color="grey")
    draw_101 = ImageDraw.Draw(image101)
    fnt = ImageFont.truetype("C:Windows/Fonts/VIVALDII.TTF", size=12)
    txt = "BATMAN"
    fnt_width, fnt_height = fnt.getsize(txt)
    
    new_width = (width - fnt_width) / 2
    new_height = (height - fnt_height) / 2
    draw_101.text((new_width, new_height), txt, fill="black")
    image101.save(output_path)

if __name__ == "__main__":
    center_text("Images/Image-Result/ImageDraw-centertext-1.jpg")

image101_1 = Image.open("Images/Image-Result/ImageDraw-centertext-1.jpg")
plt.imshow(image101_1)
```
